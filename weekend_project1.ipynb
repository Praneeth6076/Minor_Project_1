# Project Option 2: Student Marks Analyzer

```python

import numpy as np # defining numpy function
n=int(input("Enter number of students: ")) # expecting number of students input from user
marks=[]                  # creating a list named as marks
for i in range(n) :       # using loop to store the student marks in list
  m=int(input(f"Enter marks of student {i+1}: "))      # input marks from user
  marks.append(m)           # storing the input marks in list

student=np.array(marks)     # creating array of marks

total_marks=np.sum(student)       # calculating total marks of students
average_marks=np.mean(student)    # calculating average marks of students
highest_marks=np.max(student)     # calculating maximum marks of students
lowest_marks=np.min(student)      # calculating minimum marks of students

# calculating grades using conditional statements

if average_marks > 85:           # if marks are above 85
    grade = "A"                  # Grade is 'A'
elif average_marks >= 70:        # if marks are equal to or above 70
    grade = "B"                  # Grade is 'B'
elif average_marks >= 50:        # if marks are equal to or above 50
    grade = "C"                  # Grade is 'C'
else:                            # if marks are below 50
    grade = "Fail"               # student fails

print("\n-----Student marks analysis:-----\n")
print("Total Marks:", total_marks)            # printing Total marks
print("Average Marks:", average_marks)        # printing average marks
print("Highest Marks:", highest_marks)        # printing highest marks
print("Lowest Marks:", lowest_marks)          # printing lowest marks
print("Grade:", grade)                        # printing grade based on average
```
